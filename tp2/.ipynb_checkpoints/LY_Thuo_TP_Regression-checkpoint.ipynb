Nous voudrons pr√©dire le prix m√©dian des maisons californiennes par quartier en fonction de plusieurs variables quantitatives. 

Commen√ßons par charger les donn√©es et regarder la description de la base de donn√©es :


```python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)

print(california_housing.DESCR)
```

* 0.1) Montrez l'histogramme des variables (Piste : *california_housing.frame* contient un DataFrame de Pandas) 

```python
#Ecrivez votre code
import matplotlib.pyplot as plt

df = california_housing.frame

df.hist(figsize=(12, 10), bins=30, edgecolor="black")
plt.subplots_adjust(hspace=0.7, wspace=0.4)
```

## Partie I. R√©gression lin√©aire simple

Calculez la matrice de corr√©lations.

* I.1) Si vous deviez choisir une seule variable pour r√©aliser une r√©gression lin√©aire simple, laquelle choisiriez-vous ? Pourquoi ?

***R√©ponse :*** We choose variable 'MedInc' because it has a strong correlation(corr = 0.7) with target variable. 

```python
#Ecrivez votre code ici
import pandas as pd
import seaborn as sns

corrs = df.corr()
fig, ax = plt.subplots()
sns.heatmap(corrs, ax=ax, annot=True, cmap='YlGnBu', linewidths=0.2, fmt='.2f')
ax.set_title('Correlation matrix for California Housing dataset')
```

Nous voudrons cr√©er un mod√®le $ \hat{y} = ax + b $, o√π $ \hat{y} $ est la pr√©diction du prix m√©dian des maisons par quartier et $x$ est la caract√©ristique choisie en I.1.

* I.2) Quelles sont les valeurs optimales de $ a $ et $ b $ ? 

```python
#Ecrivez votre code ici
from sklearn.linear_model import LinearRegression
X = df['MedInc'].to_numpy().reshape(-1,1)
y = df['MedHouseVal']
reg = LinearRegression().fit(X,y)
print(f'a = {reg.coef_[0]}')
print(f'b = {reg.intercept_}')
```

* I.3) Quelle est la valeur de $R^2$ ? Est-ce un bon ajustement ? Commentez
* I.4) Tracez la courbe de r√©gression sur le nuage de points entre la sortie d√©sir√©e et l'entr√©e

```python
#Ecrivez votre code ici
from sklearn.metrics import r2_score
y_predict = reg.predict(X)
r2 = r2_score(y,y_predict)
print(f'Value R-squared = {r2}')

plt.scatter(X, y, color="black", s=10, label='Observations')
plt.plot(X, y_predict, color="blue", linewidth=3, label= '$\hat{y} = ax+b$')

plt.xticks(())
plt.yticks(())
plt.legend()

plt.show()
```

The value $R^2$ = 0.47 which is not a really good adjustment for the model. 

* I.5) Quelle est l'erreur RMS de pr√©diction ? (en moyenne sur toutes les observations)

```python
#Ecrivez votre code ici
import math 
import numpy as np

MSE = np.square(np.subtract(y,y_predict)).mean()

RSME = math.sqrt(MSE)

print(f'RMS error = {RSME}')
```

## Partie II. R√©gression multilin√©aire

Maintenant r√©alisez une r√©gression lin√©aire en utilisant toutes les caract√©ristiques disponibles. 

* II.1) Quelle est l'erreur de pr√©diction ?
* II.2) Quelle est la valeur de $R^2$ ? L'ajustement est-il meilleur qu'avec une seule variable d'entr√©e ?
* II.3) Quels sont les coefficients associ√©s aux variables d'entr√©e, ainsi que le terme ind√©pendant ?

```python
#Ecrivez votre code ici
from sklearn.metrics import mean_squared_error

X_multi = df.drop(['MedHouseVal'], axis=1)
reg_multi = LinearRegression().fit(X_multi,y)
print('RMSE=', mean_squared_error(y, reg_multi.predict(X_multi), squared=False))
print('R-squared =', r2_score(y, reg_multi.predict(X_multi)))

for i in range(len(reg_multi.coef_)):
    print(f'{reg_multi.feature_names_in_[i]} = {reg_multi.coef_[i]}, ')
    
print(f'Intercept = {reg_multi.intercept_}')
```

The adjustment is better than the previous model(with only 1 variable) since the vlaue of the $R^2$ is greater 

Maintenant nous allons introduire un terme de r√©gularisation, donc la nouvelle fonction de co√ªt est : $$ J' = J + \alpha\sum_{i}{w_i^2} $$
* II.4) Quelle est la valeur optimale du param√®tre $\alpha$ ? (Trouvez-la empiriquement) 
* II.5) Quelle est l'erreur RMS avec $\alpha$ optimal ?

```python
#Ecrivez votre code ici

from sklearn.linear_model import RidgeCV
ridge_cv = RidgeCV(alphas=list(range(1,1000,1))).fit(X_multi, y)
print(f'optimal_alpha = {ridge_cv.alpha_}')

Y_pred = ridge_cv.predict(X_multi)

RMSE= mean_squared_error(y, Y_pred, squared=False)

print(f'RMES_ridge = {RMSE}')
```

Ajoutez des variables polynomiales d'ordre $o=2$ aux variables d√©j√† disponibles. Comparez la r√©gression avec et sans r√©gularisation (RMSE, $R^2$)
* II.6) Laquelle donne des meilleurs r√©sultats ? Pourquoi ?  

```python
#Ecrivez votre code ici
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)

X_poly = poly.fit_transform(X_multi)

model_multi = LinearRegression().fit(X_poly, y)
Y_pred_poly = model_multi.predict(X_poly)
print("R√©gression sans r√©gularisation de variables polynomiales d'ordre 2")
RMS= mean_squared_error(y, Y_pred_poly)
print("RMSE = ", np.sqrt(RMS))
r_squared = r2_score(y, Y_pred_poly)
print("R_squared = ", r_squared)
print("----"*20)

ridge_cv = RidgeCV(alphas = list(range(1,1000,1))).fit(X_poly, y)
print(f'optimal_alpha = {ridge_cv.alpha_}')

Y_pred_poly = ridge_cv.predict(X_poly)

RMSE= mean_squared_error(y, Y_pred_poly, squared=False)
print("R√©gression avec r√©gularisation Ridge de variables polynomiales d'ordre 2")
print(f'RMES_ridge = {RMSE}')
r_squared = r2_score(y, Y_pred_poly)
print("R_squared = ", r_squared)
```

## Partie III. R√©gression non lin√©aire avec r√©seaux de neurones

Construisez et entra√Ænez un r√©seau de neurones type Perceptron Multicouche (PMC) √† $N_{layer} = 3$ couches et avec $m=5$ neurones dans la couche cach√©e.

N'oubliez pas de normaliser les donn√©es en pr√©traitement. 

```python
#Ecrivez votre code ici
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

preprocessor = StandardScaler()
norm_data = preprocessor.fit_transform(X_multi)
m = 5
mlp_model = MLPRegressor(hidden_layer_sizes=(m,m,m), max_iter=500).fit(norm_data, y)
```


* III.1) Quelle sont les prix pr√©dits pour les dix premi√®res observations ?
* III.2) Quelle est l'erreur RMS d'apprentissage ?

```python
#Ecrivez votre code ici

predict = mlp_model.predict(norm_data)
print(f'les prix pr√©dits pour les dis premi√®res observations sont {predict[:10]}')
```

```python
print('RMSE=', mean_squared_error(y, predict, squared=False))
```

R√©alisez une validation crois√©e type *K-Fold* avec $K=10$ (n'oubliez pas de m√©langer les donn√©es pr√©alablement)

* III.3) Quelle sont les erreurs RMS d'apprentissage et de test ? (Moyenne et √©cart-type)

```python
#Ecrivez votre code ici

from sklearn.model_selection import KFold, cross_validate

kf = KFold(n_splits=10, shuffle=True)
cv = cross_validate(mlp_model,norm_data, y, cv=kf, return_train_score=True)

print(f"Moyenne des performance en apprentissage : {np.mean(cv['train_score'])}")
print(f"√âcart-type des performance en apprentissage : {np.std(cv['train_score'])}")

print(f"Moyenne des performance en test: {np.mean(cv['test_score'])}")
print(f"√âcart-type des performance en test: {np.std(cv['test_score'])}")
```

Optimisez le nombre de neurones $m$ dans la couche cach√©e (Piste : r√©alisez la validation crois√©e pour diff√©rentes valeurs de $m$)

* III.4) Quelle est la valeur optimale de $m$ ? Pourquoi ? 

```python
#Ecrivez votre code ici
acc_train = []
acc_test = []
m_list = list(range(2,11))
kf = KFold(n_splits=10, shuffle=True)
for i in m_list:
    mlpmodel = MLPRegressor(hidden_layer_sizes=(i,i,i), max_iter=500).fit(norm_data, y)
    cv = cross_validate(mlpmodel, norm_data, y, cv=kf, return_train_score=True)
    acc_train.append(np.mean(cv['train_score']))
    acc_test.append( np.mean(cv['test_score']))

plt.plot(m_list,acc_train)
plt.plot(m_list,acc_test)
plt.legend(['Training Accuracy','Testing Accuracy'])
plt.title('Moyenne des performances')
plt.show()

print("Avec les performances de test:")
m_opt = m_list[acc_test.index(max(acc_test))]
print('- m_opt =', m_opt)
print('- La moyenne des performances de la validation crois√©e de m_opt =', max(acc_test))
```

Pour une validation-crois√©e type K-Fold avec $K=10$ (m√©langez l'ordre des observations), remplissez le tableau de RMSE de test ci-dessous selon les diff√©rents types de r√©gression.

| Type de r√©gression | Fold 1 | Fold 2 | Fold 3 | Fold 4 | Fold 5 | Fold 6 | Fold 7 | Fold 8 | Fold 9 | Fold 10 |
|--------------------|---------|--------|--------|-------|--------|-------|---------|--------|--------|---------|
| Lin√©aire simple    |  |  |  |  |  |  |  |  |  |  |
| Multilin√©aire sans r√©gul.    |  |  |  |  |  |  |  |  |  |  |
| Multilin√©aire avec r√©gul.    |  |  |  |  |  |  |  |  |  |  |
| Polynomiale sans r√©gul.    |  |  |  |  |  |  |  |  |  |  |
| Polynomiale avec r√©gul.    |  |  |  |  |  |  |  |  |  |  |
| Non lin√©aire avec NN    |  |  |  |  |  |  |  |  |  |  |



```python
#Ecrivez votre code ici
import pandas as pd
from sklearn.linear_model import Ridge

kf = KFold(n_splits=10, shuffle=True)

# Simple linear
reg = LinearRegression()
cv_reg = cross_validate(reg, X, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

# Multilinear
reg_multi = LinearRegression()
cv_reg_multi = cross_validate(reg_multi, X_multi, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

# Multilinear Regularization
cv_reg_ridge = cross_validate(Ridge(), X_multi, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

#Polynomial 
cv_poly = cross_validate(reg, X_poly, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

#Polynomial Regularization 
cv_poly_ridge = cross_validate(Ridge(), X_poly, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

#Neural Network 
m = m_opt
mlpmodel = MLPRegressor(hidden_layer_sizes=(m,m,m), max_iter=500)
cv_mlpmodel = cross_validate(mlpmodel, norm_data, y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

rmse_cv_reg = list(map(lambda x: -x, cv_reg['test_score']))
rmse_cv_reg_multi = list(map(lambda x: -x, cv_reg_multi['test_score']))
rmse_cv_reg_ridge = list(map(lambda x: -x, cv_reg_ridge['test_score']))
rmse_cv_poly = list(map(lambda x: -x, cv_poly['test_score']))
rmse_cv_poly_ridge = list(map(lambda x: -x, cv_poly_ridge['test_score']))
rmse_cv_mlpmodel = list(map(lambda x: -x, cv_mlpmodel['test_score']))

rmse_dict = {}
rmse_dict['Type de r√©gression']= ['Lin√©aire simple','Multilin√©aire sans r√©gul', 'Multilin√©aire avec r√©gul.', 'Polynomiale sans r√©gul.','Polynomiale avec r√©gul.','Non lin√©aire avec NN']
fold = ['Fold{}'.format(i) for i in range(1, 11)]
i = 0
for x in fold:
    rmse_dict[x] = [rmse_cv_reg[i],rmse_cv_reg_multi[i],rmse_cv_reg_ridge[i],rmse_cv_poly[i],rmse_cv_poly_ridge[i] ,rmse_cv_mlpmodel[i]]
    i = i +1

rmse = pd.DataFrame.from_dict(rmse_dict)
rmse = rmse.set_index('Type de r√©gression')
rmse
```


* III.5) Quelle est la m√©thode de r√©gression la plus adapt√©e pour ce probl√®me ? Justifiez

  ***R√©ponse:*** Le r√©seau neuronal est le meilleur mod√®le pour ce probl√®me car il donne la valeur moyenne la plus faible de RMSE de la validation crois√©e.



```python
import numpy as np

print('Mean of RMSE of simple linear:', np.mean(rmse_cv_reg))
print('Mean of RMSE of multilinear sans r√©gul:', np.mean(rmse_cv_reg_multi))
print('Mean of RMSE of multilinear avec r√©gul:', np.mean(rmse_cv_reg_ridge))
print('Mean of RMSE of polynomiale sans r√©gul:', np.mean(rmse_cv_poly))
print('Mean of RMSE of polynomiale avec r√©gul:', np.mean(rmse_cv_poly_ridge))
print('Mean of RMSE of non linear neural network', np.mean(rmse_cv_mlpmodel))
```
